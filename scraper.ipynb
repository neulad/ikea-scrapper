```python
import json
import requests as r
from bs4 import BeautifulSoup
```

```python
base = []
```

```python
response = r.get("https://www.ikea.com/ru/ru/cat/garderobnye-sistemy-paks-29879/")
html = BeautifulSoup(response.text)
sofas = html.findAll("div", "product-compact__spacer")
for sofa in sofas:
    item = {}
    children = []
    for child in list(sofa.children):
        if child != "\n":
            children.append(child)
    img = sofa.findAll("img")[0]
    item["link"] = children[0].get("href")
    item["img"] = img.get("src")
    name = list(sofa.findAll("span", attrs={"class": "product-compact__type"})[0].children)
    item["name"] = name[0].strip()
    if len(name) > 1:
        item["name"] += " " + name[1].text
    item["price"] = sofa.findAll("span", attrs={"class": "product-compact__price__value"})[0].text
    base.append(item) 
```

```python
for i in range(2, 17):
    response = r.get("https://www.ikea.com/ru/ru/cat/garderobnye-sistemy-paks-29879/" + "page-" + str(i) + "/")
    html = BeautifulSoup(response.text)
    sofas = html.findAll("div", "product-compact__spacer")
    for sofa in sofas:
        item = {}
        children = []
        for child in list(sofa.children):
            if child != "\n":
                children.append(child)
        img = sofa.findAll("img")[0]
        item["link"] = children[0].get("href")
        item["img"] = img.get("src")
        name = list(sofa.findAll("span", attrs={"class": "product-compact__type"})[0].children)
        item["name"] = name[0].strip()
        if len(name) > 1:
            item["name"] += " " + name[1].text
        item["price"] = sofa.findAll("span", attrs={"class": "product-compact__price__value"})[0].text
        base.append(item) 
with open("database_of_ikea.json", "w") as f:
    f.write("[]")
with open("database_of_ikea.json", "w") as f:
    json.dump(base, f)
```
